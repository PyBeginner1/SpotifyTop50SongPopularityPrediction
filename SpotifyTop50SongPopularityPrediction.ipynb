```python
import numpy as np
import pandas as pd

import plotly.express as px

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
song = pd.read_csv('/content/top50.csv', encoding = 'latin-1')
```

```python
song
```

```python
song.info()
```

```python
song.describe()
```

Preprocessing

```python
song = song.drop(['Unnamed: 0', 'Track.Name' ], axis = 1)
```

```python
song
```

Seperating Popularity column by less & more popular using qcut which cuts based on mean

```python
pd.qcut(song['Popularity'], q = 2, labels = ['Less Popular', 'More Popular'])
```

```python
pd.qcut(song['Popularity'], q = 2, labels = [0, 1])
```

```python
song['Popularity'] = pd.qcut(song['Popularity'], q = 2, labels = [0, 1])
```

```python
song.head()
```

**One Hot Encoding**

```python
def onehot_encode(df, column, prefix):
    df = df.copy()
    dummies = pd.get_dummies(df[column], prefix=prefix)
    df = pd.concat([df, dummies], axis=1)
    df = df.drop(column, axis=1)
    return df
```

```python
song = onehot_encode(song, 'Genre', 'genre')
song = onehot_encode(song, 'Artist.Name', 'artist')
```

```python
song.head()
```

Splitting & Scaling the data


```python
X = song.drop('Popularity', axis =1)
Y = song['Popularity']
```

```python
X
```

```python
Y
```

Scaling

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

Train & Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size =0.3, random_state = 20)
```

```python
X.shape,X_train.shape, X_test.shape
```

Model

```python
models = {
    'Logistic Regression' : LogisticRegression(),
    'K-Neighbors' : KNeighborsClassifier(),
    'Decision Tree' : DecisionTreeClassifier(),
    'Neural Network' : MLPClassifier(),
    'SVM' : SVC()
}

for name, model in models.items():
  model.fit(X_train, Y_train)
  print(name +' trained')
```

```python
for name, model in models.items():
  results = model.score(X_test, Y_test)
  print(name + ' : {:.2f}%'.format(results * 100))
```

```python
log_model = LogisticRegression()
knn_model = KNeighborsClassifier()
dec_model = DecisionTreeClassifier()
mlp_model = MLPClassifier()
svm_model = SVC()
```

```python
log_model.fit(X_train, Y_train)
knn_model.fit(X_train, Y_train)
dec_model.fit(X_train, Y_train)
mlp_model.fit(X_train, Y_train)
svm_model.fit(X_train, Y_train)
```

```python
log_acc = log_model.score(X_test, Y_test)
knn_acc = knn_model.score(X_test, Y_test)
dec_acc = dec_model.score(X_test, Y_test)
mlp_acc = mlp_model.score(X_test, Y_test)
svm_acc = svm_model.score(X_test, Y_test)
```

```python
print("Logistic Regression Accuracy: {:.2f}%".format(log_acc * 100))
print("K-Nearest-Neighbors Accuracy: {:.2f}%".format(knn_acc * 100))
print("Decision Tree Accuracy: {:.2f}%".format(dec_acc * 100))
print("Neural Network Accuracy: {:.2f}%".format(mlp_acc * 100))
print("Support Vector Machine Accuracy: {:.2f}".format(svm_acc * 100))
```

```python
fig = px.bar(
    x=["Logistic Regression", "K-Nearest-Neighbors", "Decision Tree", "Neural Network", "Support Vector Machine"],
    y=[log_acc, knn_acc, dec_acc, mlp_acc, svm_acc],
    color=["Logistic Regression", "K-Nearest-Neighbors", "Decision Tree", "Neural Network", "Support Vector Machine"],
    labels={'x': "Model", 'y': "Accuracy"},
    title="Model Accuracy Comparison"
)
fig.show()
```
